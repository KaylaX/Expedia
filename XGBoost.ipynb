```python
import pandas as pd
import datetime
from sklearn import cross_validation
import xgboost as xgb
import numpy as np
```

```python
def map10eval(preds, dtrain):
    actual = dtrain.get_label()
    predicted = preds.argsort(axis=1)[:,-np.arange(5)]
    metric = 0.
    for i in range(5):
        metric += np.sum(actual==predicted[:,i])/(i+1)
    metric /= actual.shape[0]
    return 'MAP@10', -metric
```

```python
data_xgb = pd.read_csv('data_xgb.csv')
```

```python
clf = xgb.XGBClassifier(#missing=9999999999,
                objective = 'multi:softmax',
                max_depth = 5,
                n_estimators=300,
                learning_rate=0.01,
                nthread=4,
                subsample=0.7,
                colsample_bytree=0.7,
                min_child_weight = 3,
                #scale_pos_weight = ratio,
                #reg_alpha=0.03,
                #seed=1301,
                silent=False)
```

```python
x_xgb=data_xgb.drop(['Unnamed: 0','(Intercept)','hotel_cluster'],axis=1)
y_xgb=data_xgb.iloc[:,76]
```

```python
X_train, X_test, y_train, y_test = cross_validation.train_test_split(x_xgb, y_xgb, test_size=0.2)
```

```python
xgb_test=X_test.join(y_test)
```

```python
xgb_test.shape[0]
```

```python
clf.fit(X_train, y_train, early_stopping_rounds=50, eval_metric=map10eval, eval_set=[(X_train, y_train),(X_test, y_test)])
```

```python
count = 0
chunksize = 2
preds = np.empty((xgb_test.shape[0],100))
reader = pd.read_csv('X_test_final.csv', usecols=range(1,75),chunksize=chunksize)
```

```python
for chunk in reader:
    pred = clf.predict_proba(chunk)
    preds[count:(count + chunk.shape[0]),:] = pred
    count = count + chunksize
```

```python
col_ind = np.argsort(-preds, axis=1)[:,:10]
hc = [' '.join(row.astype(str)) for row in col_ind]

sub = pd.DataFrame(data=hc)
```
